## Nightlife (Liquor Licenses)

Data collected from https://data.ny.gov/Economic-Development/Liquor-Authority-Daily-List-of-Active-Licenses-API/wg8y-fzsj with the following filter conditions:
* **County** is New York

```python
liquor_licenses = pd.read_csv('raw csvs/liquor_licenses_raw.csv')
```

```python
ops = list(liquor_licenses['Method of Operation'].value_counts().index.values)
nightlife = [x for x in ops if ("TAVERN" in x) | ("BAR" in x) | ("CLUB" in x)]
liquor_licenses = liquor_licenses[liquor_licenses['Method of Operation'].isin(nightlife)]
```

```python
liquor_licenses_cleaned = liquor_licenses.loc[:,["Method of Operation", "Premise Name", "Premise Address", "Premise City"]]
liquor_licenses_cleaned.to_csv('exported csvs/nightlife.csv')
```
